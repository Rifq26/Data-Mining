```python
# membuat variabel serta mengisi negara yang dituju
indonesia = datacovid[datacovid['location'] == 'Indonesia']
thailand = datacovid[datacovid['location'] == 'Thailand']
malaysia = datacovid[datacovid['location'] == 'Malaysia']
australia = datacovid[datacovid['location'] == 'Australia']
```

```python
# membuat variabel menentukan tanggal untuk ditetapkan serta mengisi negara yang dituju
indonesia2 = indonesia[indonesia['date'] >= '2020-01-24']
thailand2 = thailand[thailand['date'] >= '2020-01-04']
malaysia2 = malaysia[malaysia['date'] >= '2020-01-24']
australia2 = australia[australia['date'] >= '2020-01-26']
```

```python
# membuat serta mendeklarasi plot untuk setiap negara yang berisi 'date' dan 'total_deaths'
plt.plot(indonesia2['date'], indonesia2['total_deaths'])
plt.plot(thailand2['date'], thailand2['total_deaths'])
plt.plot(malaysia2['date'], malaysia2['total_deaths'])
plt.plot(australia2['date'], australia2['total_deaths'])

# menampilkan keterangan dalam grafik
plt.legend(['Indonesia', 'Thailand', 'Malaysia', 'Australia'])
#plt.xticks(rotation == 'vertical)
plt.show()
```
